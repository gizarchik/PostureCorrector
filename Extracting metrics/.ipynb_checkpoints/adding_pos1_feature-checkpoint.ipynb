```python
data = pd.read_csv('metrics_data.csv', index_col='_id')
person = pd.read_csv('../person.csv', index_col='_id')
```

### Возьмем метрики из прямого сидячего положения и добавим их к остальным метрикам (кроме самого прямого сидячего положения.

```python
added_idx = person[(person['position'] == 1) & (person['is_valid'] == 1)].index
rest_idx = person[~((person['position'] == 1) & (person['is_valid'] == 1))].index
```

```python
added_person = person.loc[added_idx].copy()
rest_person = person.loc[rest_idx].copy()
```

```python
id_cols = ['height', 'mass', 'age', 'sex']
```

```python
new_data = pd.DataFrame()
added_col_names = list(map(lambda x: x + '_pos1', data.columns))
```

```python
for added_idx, x in added_person[id_cols].iterrows():
    rest_idx = rest_person[np.all(rest_person[id_cols] == x, axis=1)].index
    
    added_row = data.loc[added_idx]
    rest_rows = data.loc[rest_idx]
    rest_rows[added_col_names] = added_row
    
    new_data = pd.concat([new_data, rest_rows])
```

```python
new_data
```

```python
rest_person = rest_person.drop(1, axis=0)
```

```python
new_data.to_csv('added_pos1_metrics_data.csv')
rest_person.to_csv('added_pos1_metrics_person.csv')
```

### Для оконных данных

```python
data = pd.read_csv('metrics_data.csv', index_col='_id')
person = pd.read_csv('../person.csv', index_col='_id')
windowed_data = pd.read_csv('windowed_metrics_data.csv', index_col=[0, 1])
```

```python
added_idx = person[(person['position'] == 1) & (person['is_valid'] == 1)].index
rest_idx = person[~((person['position'] == 1) & (person['is_valid'] == 1))].index

added_person = person.loc[added_idx].copy()
rest_person = person.loc[rest_idx].copy()
```

```python
id_cols = ['height', 'mass', 'age', 'sex']
```

```python
added_data = pd.DataFrame()
added_col_names = list(map(lambda x: x + '_pos1', data.columns))
```

```python
for added_idx, x in added_person[id_cols].iterrows():
    idx = person[np.all(person[id_cols] == x, axis=1)].index
    
    added_row = data.loc[added_idx]
    rows = data.loc[idx]
    rows[added_col_names] = added_row
    
    added_data = pd.concat([added_data, rows])
    
added_data = added_data.iloc[::,22:]
```

```python
added_data = pd.concat([added_data, person], axis=1)
```

```python
new_data = pd.concat([windowed_data.reset_index(level=1), added_data], axis=1)\
                .dropna()\
                .set_index('level_1', append=True)
```

```python
new_data
```

```python
new_data.to_csv('added_pos1_metrics_windowed.csv')
```
